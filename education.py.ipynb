```python
country_names = []
for i in soup_6.find_all(class_="tcont")[:-6]:
    country = i.find_all('td')[0].get_text()
    country_names.append(country)
    print(country)
```

```python
country_men = []
country_women = []
country_year = []
for i in soup_6.find_all(class_="tcont")[:-6]:
    men = i.find_all('td')[7].get_text()
    country_men.append(men)
    women = i.find_all('td')[10].get_text()
    country_women.append(women)
    year = i.find_all('td')[1].get_text()
    country_year.append(year)
```

```python
import sqlite3 as lite

con = lite.connect('gdpEducation.db')
cur = con.cursor()
```

```python
education_expectancy.to_csv('school_years.csv', header=True, index=False)
```

```python
with con:
    cur.execute('CREATE TABLE school_years ("country name", "year of analysis", "female school life expectancy", "male school life expectancy")')
```

```python
import csv

with open('school_years.csv') as inputFile:
    header = next(inputFile)
    inputReader = csv.reader(inputFile)
    for line in inputReader:
        to_db = [line[0], line[1], line[2], line[3]]
        with con:
            cur.execute('INSERT INTO school_years ("country name", "female school life expectancy", "male school life expectancy", "year of analysis") VALUES (?, ?, ?, ?);', to_db)
```

```python
with con:
    cur.execute('CREATE TABLE gdp (country_name text, _1999 numeric, _2000 numeric, _2001 numeric, _2002 numeric, _2003 numeric, _2004 numeric, _2005 numeric, _2006 numeric, _2007 numeric, _2008 numeric, _2009 numeric, _2010 numeric)')
```

```python
with open('GDP.csv') as inputFile:
    next(inputFile)
    next(inputFile)
    next(inputFile)
    header = next(inputFile)
    inputReader = csv.reader(inputFile)
    for line in inputReader:
        with con:
            cur.execute('INSERT INTO gdp (country_name, _1999, _2000, _2001, _2002, _2003, _2004, _2005, _2006, _2007, _2008, _2009, _2010) VALUES ("' + line[0] + '","' + '","'.join(line[43:55]) + '");')
```

```python
with con:
    cur.execute('CREATE TABLE alldata (country, gdp, "female school life expectancy", "male school life expectancy")')
```

```python
with con:
    cur.execute('INSERT INTO alldata (country, gdp, "female school life expectancy", "male school life expectancy") SELECT gdp.country_name, _2005, "female school life expectancy", "male school life expectancy" FROM gdp JOIN school_years ON gdp.country_name = school_years."country name";')
```

```python
combined_data = pd.read_sql_query("SELECT * FROM alldata", con, index_col = "country")
```

```python
combined_data
```

See if there is any correlation between the GDP numbers and the life expectancy

```python
combined_data.dropna(inplace=True) 
combined_data = combined_data[combined_data['gdp'] != '']
```

```python
combined_data
```

1 row was removed

```python
gdp = combined_data['gdp'].map(lambda x: float(x))
female_school_years = combined_data['female school life expectancy'].map(lambda x: int(x))
male_school_years = combined_data['male school life expectancy'].map(lambda x: int(x))
```

```python
import numpy as np
log_gdp = gdp.map(lambda x: np.log(x))
```

```python
plt.scatter(log_gdp, female_school_years)
plt.show()
```

```python
plt.scatter(log_gdp, male_school_years)
plt.show()
```

```python
import statsmodels.api as sm
```

```python
#Using GDP and Female School Life Expectancy

y = np.matrix(gdp).transpose()
x = np.matrix(female_school_years).transpose()

X = sm.add_constant(x)
model = sm.OLS(y,X)
results = model.fit()
results.summary()
```

```python
#Using log GDP and Female School Life Expectancy

y2 = np.matrix(log_gdp).transpose()
x2 = np.matrix(female_school_years).transpose()

X2 = sm.add_constant(x2)
model2 = sm.OLS(y2,X2)
results2 = model2.fit()

results2.summary()
```

```python
#Using GDP and Male School Life Expectancy

y3 = np.matrix(gdp).transpose()
x3 = np.matrix(male_school_years).transpose()

X3 = sm.add_constant(x3)
model3 = sm.OLS(y3,X3)
results3 = model3.fit()
results3.summary()
```

```python
#Using log GDP and Male School Life Expectancy

y4 = np.matrix(log_gdp).transpose()
x4 = np.matrix(male_school_years).transpose()

X4 = sm.add_constant(x4)
model4 = sm.OLS(y4,X4)
results4 = model4.fit()
results4.summary()
```
